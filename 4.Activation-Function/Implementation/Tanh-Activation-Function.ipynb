## **Importing Required Libraries**

```python
import numpy as np 
import matplotlib.pyplot as plt
```

## **Implementing TanH Function**

```python
# using numpy 
def tanh_activation_funx(x):
    return np.tanh(x)

x = np.linspace(-7,7,700)

tanh_activation_funx(x)
```

```python
# using mathematical equation
def tanh_activation_funx(x):
    return ((np.exp(x)- np.exp(-x))/(np.exp(x)+np.exp(-x)))

x = np.linspace(-7,7,700)
tanh_activation_funx(x)
```

```python
x = np.linspace(-10, 10, 100)
y = tanh_activation_funx(x)    
```

## ***Visualize the TanH Curve***

```python
plt.figure(figsize=(6,4))

plt.plot(x, y)
plt.title("TanH Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")

plt.axhline(0)
plt.axvline(0)

plt.grid()

plt.show()
```

## *Comparing Sigmoid vs TanH*

```python
sigmoid = 1/(1+np.exp(-x))
tanh = np.tanh(x)

plt.plot(x, sigmoid, label="Sigmoid")
plt.plot(x, tanh, label="TanH")

plt.legend()
plt.grid()
plt.show()
```
