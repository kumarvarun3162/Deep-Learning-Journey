```python
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

import tensorflow
from keras.layers import Dense
from keras.models import Sequential
```

```python
dataset = pd.read_csv(r'dataset\employee_attrition_data.csv')
```

```python
dataset.head(5)
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
dataset['Department_encoded'] = le.fit_transform(dataset['Department'])
dataset['Job_Title_encoded'] = le.fit_transform(dataset['Job_Title'])
```

```python
dataset['Department'].value_counts(), dataset['Department_encoded'].value_counts()
```

```python
dataset['Job_Title'].value_counts(), dataset['Job_Title_encoded'].value_counts()
```

```python
dataset = pd.get_dummies(dataset, columns=['Gender'], drop_first=True)

dataset['Gender_Male'] = dataset['Gender_Male'].astype(int)
```

```python
dataset.head()
```

```python
Input_data = dataset[['Age','Years_at_Company','Satisfaction_Level','Average_Monthly_Hours','Promotion_Last_5Years','Salary','Department_encoded','Job_Title_encoded','Gender_Male']]
Output_data = dataset['Attrition']
```

```python
X_train, X_test, y_train, y_test = train_test_split(Input_data,Output_data,test_size=0.2, random_state=42)
```

```python
ss = StandardScaler()

Input_data = pd.DataFrame(ss.fit_transform(Input_data),columns=Input_data.columns)
```

```python
Input_data
```

```python
ann = Sequential()
```

```python
ann.add(Dense(8, input_dim= 9, activation='relu'))
ann.add(Dense(6, activation='relu'))
ann.add(Dense(4, activation='relu'))
ann.add(Dense(2, activation='relu'))
ann.add(Dense(1, activation='sigmoid'))
```

```python
ann.compile( optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
```

```python
ann.fit(X_train,y_train, batch_size=100, epochs=70)
```

```python
pred = ann.predict(X_test)
pred_data = []
for i in pred:
    if i[0] > 0.5 : 
        pred_data.append(1)
    
    else:
        pred_data.append(0)
```

```python
accuracy_score(y_test,pred_data)*100
```

```python
pred1 = ann.predict(X_train)
pred_data1 = []
for i in pred1:
    if i[0] > 0.5 : 
        pred_data1.append(1)
    
    else:
        pred_data1.append(0)
```

```python
accuracy_score(y_train,pred_data1)*100
```
